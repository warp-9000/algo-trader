```python
import os
import requests
import pandas as pd
from dotenv import load_dotenv
import alpaca_trade_api as tradeapi
```

```python
#Load. env enviormnet variables
load_dotenv()
```

```python
#Set the varaibales for the Alpaca API Key and Secret Key 
#Type Alpaca Key and Secret Key 
alpaca_api_key= os.getenv("ALPACA_KEY")
alpaca_secret_key= os.getenv("ALPACA_SECRET_KEY")

type(alpaca_api_key)
type(alpaca_secret_key)
```

```python
#Create the Alpaca tradeapi.REST object
#Type Alpaca variable 
alpaca= tradeapi.REST(
    alpaca_api_key,
    alpaca_secret_key, 
    api_version="v2")


type(alpaca_api_key)
```

```python
#Format current date as ISO format
start_date = pd.Timestamp("2000-01-01", tz="America/New_York").isoformat()
end_date = pd.Timestamp("2022-06-30", tz="America/New_York").isoformat()
```

```python
#Set the ticker name "SPY"
tickers= ["SPY", "META", "AMZN", "AAPL","NFLX","GOOG","MSFT"]

#Set timeframe to one day ('1Day') for Alpaca API
timeframe = "1Day"
```

```python
#Get current closing prices for "SPY", "META", "AMZN", "APPL", "NFLX", "GOOG", "MSFT"
df_portfolio = alpaca.get_bars(
    tickers,
    timeframe, 
    start = start_date, 
    end = end_date 
).df


# Display sample data first 5 rows
df_portfolio.head()
```

```python
#Display sample data last 5 rows
df_portfolio.tail()
```

```python
# Separate Ticker Data for "SPY", "META", "AMZN", "APPL", "NFLX", "GOOG", "MSFT"
SPY = df_portfolio[df_portfolio['symbol']=='SPY'].drop('symbol', axis=1)
META = df_portfolio[df_portfolio['symbol']=='META'].drop('symbol', axis=1)
AMZN = df_portfolio[df_portfolio['symbol']=='AMZN'].drop('symbol', axis=1)
AAPL = df_portfolio[df_portfolio['symbol']=='AAPL'].drop('symbol', axis=1)
NFLX = df_portfolio[df_portfolio['symbol']=='NFLX'].drop('symbol', axis=1)
GOOG = df_portfolio[df_portfolio['symbol']=='GOOG'].drop('symbol', axis=1)
MSFT = df_portfolio[df_portfolio['symbol']=='MSFT'].drop('symbol', axis=1)

# Concatenate the Ticker DataFrames
df_portfolio = pd.concat(
    [SPY, META, AMZN, AAPL, NFLX, GOOG, MSFT],
    axis=1, 
    keys=['SPY','META', 'AMZN', 'NFLX', 'GOOG','MSFT']
)

# Display sample data
df_portfolio
```
